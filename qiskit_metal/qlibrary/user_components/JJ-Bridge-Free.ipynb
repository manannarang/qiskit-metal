### Josephson Junction (Bridge-Free)

The goal is to create a bridge-free Josephson Junction using BrigeFreeJunction class.  

```python
# So, let us dive right in. For convenience, let's begin by enabling
# automatic reloading of modules when they change.
%load_ext autoreload
%autoreload 2
```

```python
import numpy as np
import qiskit_metal as metal
from qiskit_metal import Dict, draw
from qiskit_metal.qlibrary.core import QComponent

```

Import the class from QComponent library. 

from qiskit_metal.qlibrary.qubits.JJ_Brige_Free import BrigeFreeJunction

```python
#on my own pc
from brige_free_jj import BridgeFreeJunction
```

```python
design=metal.designs.DesignPlanar()

design.overwrite_enabled = True

#Launch Qiskit Metal GUI 
gui = metal.MetalGUI(design)
```

```python
#Let's create one junction at position (0,0)
bf_J1 = BridgeFreeJunction(design, 'J1', options = dict(pos_x = '-0.5 mm', pos_y = '+1.0 mm'))


gui.rebuild()
gui.autoscale()
gui.zoom_on_components(['J1'])
gui.screenshot()

```

```python
# Screenshot the canvas only as a .png formatted file.
gui.figure.savefig('shot.png')

from IPython.display import Image, display
_disp_ops = dict(width=500)
display(Image('shot.png', **_disp_ops))
```

If we want to rotate it to an arbitrary angle: 

```python
bf_J2 = BridgeFreeJunction(design, 'J2', options = dict(pos_x = '-0.4 mm', pos_y = '+1.0 mm', orientation = 45))


gui.rebuild()
gui.autoscale()
gui.zoom_on_components(['J2'])
gui.screenshot()

```
